```python
import os
import json
import csv
import shutil
import pandas as pd
```

```python



class Load_Data:
    def __init__(self, data_path, new_path):
        self.data_path  = data_path
        self.new_path = new_path

    def load_data(self):
        for folder in os.listdir(self.data_path):
            print(folder)
            os.makedirs(os.path.join(self.new_path, folder), exist_ok=True)
            for file in os.listdir(os.path.join(self.data_path, folder)):
                if file.endswith(".csv"):
                    shutil.copy(os.path.join(self.data_path, folder, file), os.path.join(self.new_path, folder, file))
                if file.endswith(".bmp"):
                    shutil.copy(os.path.join(self.data_path, folder, file), os.path.join(self.new_path, folder, file))
                if file.endswith(".json"):
                    os.remove(os.path.join(self.new_path, folder, file))    
    def make_labelme_template(self):
        for folder in os.listdir(self.new_path):
            print(folder)
            for file in os.listdir(os.path.join(self.new_path, folder)):
                if file.endswith(".csv"):
                    with open(os.path.join(self.new_path, folder, file), 'r') as f:
                        csv_reader = pd.read_csv(f)
                        shapes = []
                        for index, row in csv_reader.iterrows():
                            shape = {
                                "label": "numb_box",
                                "points": [
                                    [row['Top_Left_X'], row['Top_Left_Y']],
                                    [row['Bottom_Right_X'], row['Bottom_Right_Y']],
                                    [row['Top_Right_X'], row['Top_Right_Y']],
                                    [row['Bottom_Left_X'], row['Bottom_Left_Y']],
                                ],
                            "group_id": None,
                            "description": "",
                            "shape_type": "polygon",
                            "flags": {},
                            "mask": None          
                            }
                            shapes.append(shape)

                        data = {
                            "version": "5.5.0",
                            "flags": {},
                            "shapes": shapes,
                            "imagePath": file[:3] + ".bmp",
                            "imageData": None,
                            "imageHeight": 3648,
                            "imageWidth": 5472
                        }

                        with open(os.path.join(self.new_path, folder, file[:3] + ".json"), 'w') as f:
                            json.dump(data, f, indent=4) 
    
                            
def main():
    path = r"D:\Work\codev2\code\work_code\preprocess_data\fix"
    new_path = r"D:\Work\codev2\code\work_code\preprocess_data\fix"
    data = Load_Data(path, new_path)
    # data.load_data()
    data.make_labelme_template()
main()
```
