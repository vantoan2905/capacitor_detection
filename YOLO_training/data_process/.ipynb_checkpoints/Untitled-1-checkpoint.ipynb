```python
import numpy as np
from PIL import Image
```

```python
import matplotlib.pyplot as plt
def viz(img):
    
    # plt.figure(figsize=(15, 8))
    plt.imshow(img)
```

```python
def split_image_with_overlap(image_path, num_rows=2, num_cols=3, overlap_percent=0.2):
    """
    Chia hình ảnh thành nhiều phần có overlap.
    
    :param image_path: Đường dẫn đến hình ảnh
    :param num_rows: Số hàng để chia (mặc định là 2)
    :param num_cols: Số cột để chia (mặc định là 3)
    :param overlap_percent: Phần trăm overlap giữa các phần (mặc định là 20%)
    :return: Danh sách các phần hình ảnh và tọa độ của chúng
    """
    image = Image.open(image_path)
    width, height = image.size
    
    # Tính toán kích thước của mỗi phần và overlap
    part_width = width // num_cols
    part_height = height // num_rows
    overlap_x = int(part_width * overlap_percent)
    overlap_y = int(part_height * overlap_percent)
    
    splits = []
    for i in range(num_rows):
        for j in range(num_cols):
            # Tính toán tọa độ cho phần hiện tại
            x1 = max(0, j * part_width - overlap_x)
            y1 = max(0, i * part_height - overlap_y)
            x2 = min(width, (j + 1) * part_width + overlap_x)
            y2 = min(height, (i + 1) * part_height + overlap_y)
            
            # Cắt phần hình ảnh
            part = image.crop((x1, y1, x2, y2))
            print((x1, y1, x2, y2))
            viz(part)
            splits.append((part, (x1, y1, x2, y2)))
    
    return splits
```
